```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
UserStats = json.load(open(r'./Desktop/UserStats.txt'))
```

```python
TotalMessageDist = []
TotalThreadDist = []
CombinedDomainDict = {}


for UserID in UserStats:
    
    TotalMessageDist.append(UserStats[UserID]['TotalMessages'])
    TotalThreadDist.append(UserStats[UserID]['TotalThreads'])
    
    for DomainName in UserStats[UserID]['DomainDist']:
        
        if DomainName in CombinedDomainDict:
            CombinedDomainDict[DomainName] += UserStats[UserID]['DomainDist'][DomainName]
        else:
            CombinedDomainDict[DomainName] = UserStats[UserID]['DomainDist'][DomainName]
```

```python
TotMessageDist = pd.Series(TotalMessageDist)

TotMsgDist = plt.gcf()
TotMessageDist.plot(kind='hist', bins=50, normed=False, color='magenta', grid=True)
plt.ylim([0, 1550])
plt.xlim([0, 60000])
plt.xlabel('Total Number of Emails')
plt.ylabel('Number of Users')
plt.title('Full Scale')

TotMsgDist.savefig(r'.\Desktop\TotalMessageDist_FullScale.png')
```

```python
TotMsgDist = plt.gcf()
TotMessageDist.plot(kind='hist', bins=50, normed=False, color='magenta', grid=True)
plt.ylim([0, 500])
plt.xlim([0, 60000])
plt.xlabel('Total Number of Emails')
plt.ylabel('Number of Users')
plt.title('Zoomed')

TotMsgDist.savefig(r'.\Desktop\TotalMessageDist_Zoomed.png')
```

```python
TotThreadDist = pd.Series(TotalThreadDist)

TotThrDist = plt.gcf()
TotThreadDist.plot(kind='hist', bins=50, normed=False, color='cyan', grid=True)
plt.ylim([0, 1550])
plt.xlim([0, 60000])
plt.xlabel('Total Number of Threads')
plt.ylabel('Number of Users')
plt.title('Full Scale')

TotThrDist.savefig(r'.\Desktop\TotalThreadDist_FullScale.png')
```
