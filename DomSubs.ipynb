```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
DomainSubjectCounts = json.load(open(r'./Desktop/DomainSubjectCounts.txt'))
```

```python
DomainSubjectInfo = pd.DataFrame(columns=['Domain', 'Subject', 'Count'])
i = 0

for Domain in DomainSubjectCounts.keys():
    for Subject in DomainSubjectCounts[Domain]:
        DomainSubjectInfo.loc[i] = [Domain, Subject, DomainSubjectCounts[Domain][Subject]]
        i += 1
```

```python
DomainSubjectInfo.set_index(['Domain'], inplace=True)
```
